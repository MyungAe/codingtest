```python
## .sort() 메서드 (w/ key)
numbers = [5, 6, 1, 3, 2, 4]
numbers.sort()  # 오름차순 정렬
print(numbers)
numbers.sort(reverse=True)  # 반대로 => 내림차순 정렬
print(numbers)

balances = [[24, 100], [12, 100000], [35, 6000], [3, 10e8]]

balances.sort(key=lambda elem: elem[0], reverse=True)
# 계좌 잔액 많은 순으로
balances.sort(key=lambda e: e[1], reverse=True)
print(balances)
```

## 반복

### `while`

```python
# while 5번 출력하기
n = 0
while n < 5:
    print('Five Times')
    n += 1
```

```python
# while로 numbers 모두 더하기
numbers = [1, 2, 3, 4, 5]

idx = 0
total = 0
while idx < len(numbers):
    total += numbers[idx]
    idx += 1

total
```

### `for`

```python
# for로 numbers 모두 더하기
numbers = [1, 2, 3, 4, 5]

total = 0
for num in numbers:
    total += num

total
```

```python
# for로 5번 출력하기
for _ in range(5):  # 변수명이 _일 경우 => 무시해라
    print('Five Times')
```

```python
# for로 string 순회하기
for char in 'python':
    print(f'!{char}!')
```

### 직접 함수 만들기

```python
# 여러번 인사하는 함수 만들기
def hello_n_times(n):
    for i in range(n):
        print(f'hello => {i}')

hello_n_times(5)  # => 5번 인사말이 출력 되고 싶다..
```

```python
# 원 면적 구하는 함수 만들기
def get_circle_area(r):
    pi = 3.141592
    area = pi * r**2
    return area

get_circle_area(10) # => 반지름이 10인 원의 넓이 구하고 싶다.
# DRY => Don't Repeat Yourself

get_circle_area(3) + get_circle_area(5)
```

```python
## lambda 표현식
def a(x):  # 기명 함수
    return x**2

b = lambda x: x ** 2

print(a(2), b(5))

# 익명 함수
# 1. def 지우고 lambda 적고
# 2. 함수 이름 지우고
# 3. 소괄호 지우고
# 4. 엔터 지우고
# 5. return 지운다

(lambda x: x ** 2)(3)
```
